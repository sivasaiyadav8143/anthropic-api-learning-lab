```python
# Chunk by a set number of characters
def chunk_by_char(text, chunk_size=150, chunk_overlap=20):
    """
    Split `text` into chunks of up to `chunk_size` characters, with `chunk_overlap` characters
    shared between consecutive chunks.

    Parameters:
    - text (str): Input string to chunk.
    - chunk_size (int): Maximum number of characters per chunk.
    - chunk_overlap (int): Number of characters to overlap between consecutive chunks.

    Returns:
    - List[str]: A list of string chunks.

    Notes:
    - Overlap helps maintain context between chunks when processing long text with models.
    - If the final chunk is shorter than `chunk_size`, the loop ends when the end of the text is reached.
    """
    chunks = []
    start_idx = 0

    # Continue slicing until we've covered the entire text
    while start_idx < len(text):
        # Determine the end index for this chunk (don't go past end of text)
        end_idx = min(start_idx + chunk_size, len(text))

        # Extract the chunk and append to results
        chunk_text = text[start_idx:end_idx]
        chunks.append(chunk_text)

        # Advance start index so we create overlap between chunks. If we've reached the text end,
        # set start_idx to len(text) to break the loop.
        start_idx = (
            end_idx - chunk_overlap if end_idx < len(text) else len(text)
        )

    return chunks
```

```python
# Chunk by sentence
import re


def chunk_by_sentence(text, max_sentences_per_chunk=5, overlap_sentences=1):
    """
    Split `text` into chunks by sentences. Each chunk will contain up to
    `max_sentences_per_chunk` sentences and consecutive chunks will share
    `overlap_sentences` sentences to preserve context across boundaries.

    Parameters:
    - text (str): Input text to split into sentence-based chunks.
    - max_sentences_per_chunk (int): Maximum sentences per chunk.
    - overlap_sentences (int): Number of sentences to overlap between chunks.

    Returns:
    - List[str]: A list of chunk strings where sentences in each chunk are joined by a space.

    Implementation notes:
    - Uses a simple regex to split on sentence-ending punctuation followed by whitespace.
    - The overlap helps retain continuity when analyzing or embedding chunks.
    - Edge cases: If `overlap_sentences` >= `max_sentences_per_chunk`, the start index
      progression may stall; ensure sensible parameter values.
    """
    # Split text into sentences using punctuation-based boundary detection
    sentences = re.split(r"(?<=[.!?])\s+", text)

    chunks = []
    start_idx = 0

    # Advances by (max_sentences_per_chunk - overlap_sentences) to create overlapping windows
    while start_idx < len(sentences):
        end_idx = min(start_idx + max_sentences_per_chunk, len(sentences))

        current_chunk = sentences[start_idx:end_idx]
        # Join the selected sentences back into a single text chunk
        chunks.append(" ".join(current_chunk))

        # Move the window forward, leaving `overlap_sentences` at the end of the previous window
        start_idx += max_sentences_per_chunk - overlap_sentences

        # Defensive: ensure start_idx never becomes negative
        if start_idx < 0:
            start_idx = 0

    return chunks
```

```python
# Chunk by section using markdown-style headings

def chunk_by_section(document_text):
    """
    Split a document into sections using markdown '## ' headings as delimiters.

    Parameters:
    - document_text (str): Full document text to split.

    Returns:
    - List[str]: Sections of the document. The first element may contain text
      that appears before the first '## ' heading.

    Note: This relies on a simple pattern and is suitable for documents that
    use '## ' to denote section headings. For different heading styles, adjust
    the `pattern` accordingly.
    """
    pattern = r"\n## "
    return re.split(pattern, document_text)
```

```python
# Read a markdown report and chunk it using the chosen chunking function
with open("./report.md", "r", encoding="utf-8") as f:
    text = f.read()

# Create character-based chunks using the default chunk_by_char parameters
chunks = chunk_by_char(text)

# Print each chunk separated by a visible delimiter for readability
for chunk in chunks:
    print(chunk + "\n----\n")
```
